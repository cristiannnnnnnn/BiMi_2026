# TF-IDF + Bayes -- train

```python
import pandas as pd

data = pd.read_csv('../sentences_top100.csv').dropna(subset=['content']) # remove every line with null value
data = data[data['content'].str.strip() != ''].reset_index(drop=True) # remove any remaining empty strings
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data['content'],
    data['lecturer_id'],
    test_size=0.2,
    random_state=42,
    stratify=data['lecturer_id'], #ensure that every lecturer has 80/20 split in train and test
)

len(X_train), len(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),     # match single words (unigrams) and pairs of words (bigrams)
    min_df=5,               # ignore terms that appear in less than 5 sentences
    max_df=0.5,             # ignore terms that appear in more than 50% of the sentences
    max_features=50_000,    # avoid 1h vectorization runtime
)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_train_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(X_train_tfidf, y_train)
```

```python
import joblib

joblib.dump(vectorizer, 'vectorizer.joblib')
joblib.dump(clf, 'model.joblib')
print('saved vectorizer.joblib and model.joblib')
```
